### 3. **DataFrame.explode()** This method is useful for reshaping a list inside a dataframe into multiple rows.


**References **: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.explode.html

```python
dfExplode = pd.DataFrame(data={"Phones": ['Samsung', 'Iphone'], 
                        "Versions": [['Galaxy A22','Galaxy Z Fold' ,'Galaxy Z Flip'], ['12 Pro', '11', 'XR']]})

#regular display
print(dfExplode)

#explodes display  and creates a new index as it resizes, instead of the index being just 0 and 1 as it is in the original
dfExplode.explode("Versions", ignore_index = True)
```
