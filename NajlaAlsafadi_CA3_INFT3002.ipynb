**Why is loading big data in batches an effective startegy?**

It also provides a higher data quality with better precision and accuracy due to the batches.

This happens because the model can't capture complexity or there is a lack of relationships between the data. Model with high bias doesn't take into account detail well and oversimplifies it.

## **NN and Parameter Selection**

To explain the code below:
Firstly I get the wine data set and train on model, its not reliable so I then split the data into 0.3. I will refer to the example used in class that similar to a normal test, we practice more questions and the actual test will be less. So you can see we train for 124 and test on 54.

Then use 1 nn and fit data and make predictions and compare against y test true predicitons



```python
from sklearn import datasets
wine = datasets.load_wine()
X = wine.data
y = wine.target
#print(wine['DESCR'])
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(wine['data'], wine['target']) #train on model
print(wine['data'].shape)
print(wine['target'].shape)
#not fitting and not reliable
```

```python
#iinstead we split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21, stratify=y)#stratify means if unequal make equal

print('train', X_train.shape, y_train.shape)
print('test', X_test.shape, y_test.shape)
# train for 124 and test on 54
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
print("Test set true labels:\n {}".format(y_test))
print('Accuracy', knn.score(X_test, y_test))
```

Above you can see the two result set for the computer result nd the actual result. The results show that the data set predictions isn't the exact same true labels as there isn't guaruantee in full accuarcy. So we use accuarcy to see how many times the computer got it correct versus the actual value.

Below you can see what k score the testing and trainign accuracy is the highest. Using visualisation is better to display results. Range between 1 to 10, every time it runs for each number fit x and y train and get score calculate the test score which is the real score. 

```python
import numpy as np
# Setup arrays to store train and test accuracies
neighbors = np.arange(1, 10)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

# Loop over different values of k
for i, k in enumerate(neighbors):
    # Setup a k-NN Classifier with k neighbors: knn
    knn = KNeighborsClassifier(n_neighbors=k)

    # Fit the classifier to the training data
    knn.fit(X_train, y_train)

    #Compute accuracy on the training set
    train_accuracy[i] = knn.score(X_train, y_train)

    #Compute accuracy on the testing set
    test_accuracy[i] = knn.score(X_test, y_test)

# Generate plot
plt.title('k-NN: Varying Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

Now we can say that nn of 7 will get the highest accuracy based on the graph and now i changed the value to display the value equal to 7 shwoing the accuracy to be 0.79 which is higher than nn=1 at 0.74. I used 7 as opposed to 8 as its not perfered to use even numbers in such a case as discussed in class. **This shows the influence of selecting specific parameter values**

```python
knn = KNeighborsClassifier(n_neighbors=7) # use 7 nn based of above data and fit data and make predictions and compare against y test true predicitons
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred))
print("Test set true labels:\n {}".format(y_test))
print('Accuracy', knn.score(X_test, y_test))
```

## **NC and Parameter Selection and Visualizing Decision Boundary**

How datasets looks in a 2d plane for all the points and the centorid is the star for each colour. The **decision boundry** is the divide in colour and easy to see becuase its only split into 3.

For **parameter tuning** we look at the shrinkage and see results at none(which is default essentally) and .2, 1.2, 1.5. And display the results at each point. The result for each point is quite similar in this case. its possible there could be different shrinkage values that could be used.  

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.neighbors import NearestCentroid



# import some data to play with
wine = datasets.load_wine()
# we only take the first two features. We could avoid this ugly
# slicing by using a two-dim dataset
X = wine.data[:, :2]
y = wine.target

h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

for shrinkage in [None, .2, 1.2, 1.5]:
    # we create an instance of Neighbours Classifier and fit the data.
    nc = NearestCentroid(shrink_threshold=shrinkage)
    nc.fit(X, y)
    y_pred = nc.predict(X)
    print(shrinkage, np.mean(y == y_pred))
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = nc.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20, alpha=0.65)
    plt.scatter(nc.centroids_[:, 0], nc.centroids_[:, 1], c=[0, 1, 2], cmap=cmap_bold,
                edgecolor='k', s=160, marker='*')
    plt.title("3-Class classification (shrink_threshold=%r)"
              % shrinkage)
    plt.axis('tight')

plt.show()
```

## **Computational Comaparison**

```python
# time to build the model, memory to store the model, and time to make a prediction.

def vcopies(arr, num=4):
    return np.vstack( (arr for i in range(num)) )

def hcopies(arr, num=4):
    return np.hstack( (arr for i in range(num)) )


X_train2 = vcopies(X_train)
y_train2 = hcopies(y_train)
X_test2 = vcopies(X_test)
y_test2 = hcopies(y_test)

def basic_knn():
    knn = KNeighborsClassifier(n_neighbors=1, algorithm="kd_tree")
    knn.fit(X_train2, y_train2)
    y_pred2 = knn.predict(X_test2)
    print("Test set predictions:\n {}".format(y_pred2))
    print(knn.score(X_test2, y_test2))

def basic_nc():
    nc = NearestCentroid()
    nc.fit(X_train2, y_train2)
    display(nc.centroids_)
    print("accuracy: {:.2f}".format(nc.score(X_test2, y_test2)))

basic_knn()
basic_nc()
```

```python
# Performance


import timeit
number = 2
repeat = 4
knn_train_time = timeit.repeat(
    'knn = KNeighborsClassifier(n_neighbors=1, algorithm="kd_tree"); knn.fit(X_train2, y_train2)',
     globals=globals(), number=number, repeat=repeat)

nc_train_time = timeit.repeat(
    'nc = NearestCentroid(); nc.fit(X_train2, y_train2)',
     globals=globals(), number=number, repeat=repeat)

knn_predict_time = timeit.repeat(
    'knn.predict(X_test2)',
     globals=globals(), number=number, repeat=repeat)

nc_predict_time = timeit.repeat(
    'nc.predict(X_test2)',
     globals=globals(), number=number, repeat=repeat)


def get_avg_time(time_take):
    return np.array(time_take).mean()/number
    
knn_train_avg_time = get_avg_time(knn_train_time)
knn_predict_avg_time = get_avg_time(knn_predict_time)
nc_train_avg_time = get_avg_time(nc_train_time)
nc_predict_avg_time = get_avg_time(nc_predict_time)
print('knn', knn_train_avg_time)
print('nc',nc_train_avg_time)
print('knn',knn_predict_avg_time)
print('nc',nc_predict_avg_time)
```

To see how the nc vs nn perform under time I used the above code, the time difference and can see NC is better. It's not easy to see from the above values so I displayed the results to see **visual computaional comparison**
